```python
import numpy as np 
import tensorflow as tf 
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense, Input 
from tensorflow.keras import optimizers 
import matplotlib.pyplot as plt
# Set seed for reproducibility
seed = 0
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
# Load data
raw_data = np.loadtxt('data/heart.csv', delimiter=',', skiprows=1)
x_data = np.array(raw_data[:, 0:13], dtype=np.float32)
y_data = np.array(raw_data[:, 13], dtype=np.float32)
y_data = y_data.reshape(1025, 1) # Reshape
# Build model
model = Sequential()
model.add (Input(shape=(13,))) # Use Input object to define input shape 
model.add(Dense(32, activation='relu')) 
model.add(Dense(16, activation='relu')) 
model.add(Dense(1, activation='sigmoid'))
# Compile model
opt = optimizers.Adam(learning_rate=0.01)
model.compile(loss='binary_crossentropy', optimizer=opt, metrics=['accuracy'])
# Train model
history = model.fit(x_data, y_data, epochs=500, batch_size=50, verbose=2)
```

```python
# Evaluate model.
accuracy = model.evaluate(x_data, y_data, verbose=0) [1]
print('\n Accuracy: %.4f' % accuracy)
# Plotting accuracy and loss
plt. figure(figsize=(12, 5))
# Plot accuracy
plt. subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.title( 'Model Accuracy')
plt.xlabel( 'Epoch')
plt.ylabel( 'Accuracy')
plt. grid (True)
# Plot loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.title('Model Loss')
plt.xlabel( 'Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
new_data = np.array ([[47, 1, 0, 110, 275, 0, 0, 118, 1, 1, 1, 1, 21]], dtype=np.float32)
```

```python
# Make prediction
prediction = model.predict(new_data)
print(f"Predicted value: {prediction[0][0]}")
```
