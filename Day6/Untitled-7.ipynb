케이폴드 교차검증

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold

import pandas as pd
import numpy as np
```

```python
seed = 0
np.random.seed(seed)
tf.random.set_seed(seed)
df =pd.read_csv('data/sonar.csv', header=None)
dataset = df.values
X = dataset[:, 0:60]

Y_obj = dataset[:, 60]
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
```

```python
n_fold = 10
skf = StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=seed)
```

```python
accuracy = []

for train, test in skf.split(X, Y):
    model = Sequential()
    model.add(Dense(24, input_dim=60, activation='relu'))
    model.add(Dense(10, activation='relu')) 
    model.add(Dense(1, activation='sigmoid'))
    model. compile(loss='mean_squared_error' ,optimizer='adam' ,metrics=['accuracy'])
    ##× = dataset[:, 0:60].astype(float)를 쓰면 아래 두줄 생략 가능
    
    X_train = tf.convert_to_tensor(X[train], dtype=tf.float32)
    Y_train = tf. convert_to_tensor(Y[train], dtype=tf.float32)
    model. fit(X_train, Y_train, epochs=100, batch_size=5)
    X_test = tf.convert_to_tensor(X[test], dtype=tf.float32)
    Y_test = tf.convert_to_tensor(Y[test], dtype=tf.float32)
    k_accuracy = "%.4f" % (model.evaluate(X_test, Y_test) [1])
    accuracy.append (k_accuracy)
```

```python
print("\n %.f fold accuracy:" % n_fold, accuracy)
```
