```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
import numpy as np
```

```python
data = np.loadtxt("ThoraricSurgery.csv", delimiter=',')
x_data = data[:,0:-1]
y_data = data[:,[-1]]
```

```python
x_data.shape
```

```python
X = tf.placeholder(tf.float32, [None, 17])
Y = tf.placeholder(tf.float32, [None, 1])
```

```python
learning_rate = 0.1
tf.set_random_seed(0)
np.random.seed(0)
```

```python
W1 = tf.Variable(tf.random_normal([17, 30]), name='weight1')
b1 = tf.Variable(tf.random_normal([30]), name='bias1')
layer1 = tf.sigmoid(tf.matmul(X, W1) + b1)

W2 = tf.Variable(tf.random_normal([30, 1]), name='weight2')
b2 = tf.Variable(tf.random_normal([1]), name='bias2')
hypothesis= tf.sigmoid(tf.matmul(layer1, W2) + b2)
```

```python
cost = -tf.reduce_mean(Y * tf.log(hypothesis) 
                      + (1 - Y) * tf.log(1 - hypothesis))
train = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)

predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y),
                                  dtype=tf.float32))
```

```python
new_data = np.array ([[132, 2,2.12, 1.72, 1, 0, 0, 0,0,0, 12, 0, 0, 0, 1,0, 74]], dtype=np.float32)
```

```python
with tf.Session() as sess:
    # TensorFlow variables 
    sess.run(tf.global_variables_initializer())
    for step in range(10001):
        sess.run(train, feed_dict={X: x_data, Y: y_data})
        
        if step % 100 == 0:
            print("step = ", step, 
                  "cost = ", sess.run(cost, feed_dict={X: x_data, Y: y_data}),
                  "W = ", sess.run(W2))
    # Accuracy report
    h, c, a = sess.run([hypothesis, predicted, accuracy], 
                       feed_dict={X: x_data, Y: y_data})
    print("\nHypothesis: ", h, "\nCorrect: ", c, "\nAccuracy: ", a)

    new_prediction = sess.run(hypothesis, feed_dict={X: new_data})
    new_predicted_class = sess.run(predicted, feed_dict={X: new_data})

    print("\nNew Data Prediction (Probability): ", new_prediction)
    print("New Data Predicted Class: ", new_predicted_class)
```
