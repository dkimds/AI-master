```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense 
import numpy
import tensorflow as tf
seed = 0
numpy.random.seed (seed)
tf.random.set_seed (seed)
```

```python
Data_set = numpy.loadtxt("data/ThoraricSurgery.csv", delimiter=',')
```

```python
X= Data_set[:, 0:17]
Y = Data_set[:, 17]
```

```python
model = Sequential()
model.add(Dense(30, input_dim=17, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy' , optimizer='adam', metrics=['accuracy'])
model. fit(X, Y, epochs=30, batch_size=10)
```

```python
print("in Accuracy: %.4f" % (model.evaluate(X, Y)[1]))
```

```python
from tensorflow.keras import optimizers
```

```python
import numpy as np
```

```python
x_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]],'float32')
y_data = np.array([[0], [1], [1], [0]],'float32')
```

```python
model = Sequential()
model.add(Dense(4, input_dim=2, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
```

```python
opt = optimizers.Adam(learning_rate=0.1)
```

```python
model.compile(loss='binary_crossentropy' , optimizer=opt, metrics=['accuracy'])
model.fit(x_data, y_data, epochs=1000, verbose=2)
```

```python
print("in Accuracy: %.4f" % (model.evaluate(x_data, y_data)[1]))
```

다중 선형 회귀 케라스

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
import numpy as np
from tensorflow.keras.optimizers import RMSprop
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

```

```python
raw_data = np.loadtxt("../data/Blood_fat.csv", delimiter=',')
```

```python
xs = np.array(raw_data[:, 0], dtype=np.float32)
ys = np.array(raw_data[:, 1], dtype=np.float32)
zs = np.array(raw_data[:, 2], dtype=np.float32)
x_data = np.array(raw_data[:, 0:2], dtype=np.float32)
y_data = np.array(raw_data[:, 2], dtype=np.float32)
y_data = y_data.reshape(25, 1)
```

```python
rmsprop = RMSprop(learning_rate=0.01)
model = Sequential()
```

```python
model.add(Dense(1, input_shape=(2,)))
model.compile(loss='mse' , optimizer=rmsprop)
model.summary()
```

```python
hist = model.fit(x_data, y_data, epochs=1000)
```

```python
print(hist.history.keys())
print('100kg  40세 혈중지방함량치=', model.predict(np.array([100, 40]).reshape(1,2)))
print('60kg  25세 혈중지방함량치=', model.predict(np.array([60, 25]).reshape(1,2)))
```

graph

```python
W_,b_ = model.get_weights()
```

```python
x = x_data[:,0]
y = x_data[:,1]
Z = np.matmul(x_data, W_) + b_
```

```python
x_l = np.linspace(np.min(x_data,axis=0)[0],
        np.max(x_data,axis=0)[0], len(x_data)).reshape(25,1)
y_l = np.linspace(np.min(x_data,axis=0)[1],
        np.max(x_data,axis=0)[1], len(x_data)).reshape(25,1)
print ('x=',x_l.shape)
print('y=',y_l.shape)
```

```python
X = np.concatenate((x_l, y_l), axis=1)
print("X=", X.shape)
Z_l = np.matmul(X, W_) + b_
```

```python
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111,projection='3d')

Z_reshaped = Z.reshape(25)
ax.scatter(x,y,Z_reshaped, label='Predict')

ax.scatter(xs, ys, zs, label="Real")

ax.scatter(x_l, y_l, Z_l, label='Linear')

ax.set_xlabel('weight')
ax.set_ylabel('age')
ax.set_zlabel('blood fat')
ax.view_init(15, 15)
plt.legend(loc='upper center')
plt.show()
```
