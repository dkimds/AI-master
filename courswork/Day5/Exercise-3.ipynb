케라스 익히기 스몰 프로젝트1

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
import numpy as np
from tensorflow.keras.optimizers import RMSprop
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x_train = np.array([-12, -10,-8,-6, -4, -2, 2, 4, 6, 8, 10,12])
y_train = np.array([144, 100, 64, 36, 16, 4, 4, 16, 36, 64, 100, 144])
```

```python
rmsprop = RMSprop(learning_rate=0.01)
model = Sequential()
model.add(Dense(20, input_dim=1, activation='relu'))
model.add(Dense(1, input_dim=20))

model.compile(loss='mse' , optimizer=rmsprop)
model.summary()
```

```python
hist = model.fit(x_train, y_train, epochs=1000, batch_size=4)
```

```python
new_x = [-11, -5, -3, 3, 5, 11]
for i in new_x:
    print(f'Prediction of {i} squre is', float(model.predict(np.array(i).reshape(1,), verbose=0)))
```

```python
x = x_train[:]
y = y_train[:]

Z = model.predict(np.array(new_x))
```

```python
print(x)
print(y)
print(Z)
```

```python
Z_reshaped = Z.reshape(6)
```

```python
plt.title("actual vs predicted")
plt.plot(x, y, 'r--')
plt.scatter(new_x, Z_reshaped)
plt.show()
```
